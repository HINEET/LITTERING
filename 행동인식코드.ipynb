```python
import cv2
from cvzone.PoseModule import PoseDetector
import base64
from IPython.display import HTML
from base64 import b64encode

# Set up video capture
cap = cv2.VideoCapture("/content/drive/MyDrive/빅프로젝트관련파일/model_test_file/litteringtestvideo/8.mp4")
detector = PoseDetector()
posList = []

# Prepare video writer
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter('carhuman1.mp4', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

while True:
    ret, img = cap.read()
    if not ret:
        break

    # Detect pose
    img = detector.findPose(img)
    lmList, _ = detector.findPosition(img, bboxWithHands=False)

    # Check if landmarks were detected and process them
    if lmList:
        lmString = ''
        for lm in lmList:
            if len(lm) >= 4:  # Check if landmark data is complete
                lmString += f'{lm[1]},{img.shape[0] - lm[2]},{lm[3]},'
        posList.append(lmString)

    # Write frame to output video file
    out.write(img)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit the video stream
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()

# Display video in Colab
mp4 = open('carhuman1.mp4','rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML("""
<video width=400 controls>
      <source src="%s" type="video/mp4">
</video>
""" % data_url)
```
